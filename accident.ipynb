```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
df = pd. read_csv("driverresponse.csv")
df. info
df. head()
```

```python
df = df. iloc[:,2:]
df. head()
```

```python
#Data Visualization
for column in df.columns:
    plt.hist(df[column])
    plt.title(column)
    plt.show() 
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
sns.pairplot(df.iloc[:, :])
```

```python
df = df. drop(["regionid"], axis = 1)
df. head()
```

```python
df['Total'] = df. sum(axis = 1)
df. head()
```

```python
#Scatterplot
plt. scatter(x = df["overspeed2014"], y = df["Total"], color = 'green')
```

```python
#Scatterplot
plt. scatter(x = df["alcintake2014"], y = df["Total"], color = 'green')
```

```python
df_region = df. groupby('region', axis=0).sum()
print(type(df. groupby('region', axis=0)))
df_region. head()
```

```python
df_region['Total'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,    
                            )

plt.title('Total accident in region')
plt.axis('equal')
plt.show()
```
